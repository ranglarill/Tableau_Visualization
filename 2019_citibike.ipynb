```python
import pandas as pd
import glob
```

```python
path = r'Resources/2019' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
li = []
```

```python
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
```

```python
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
new_frame = frame[(frame != 0).all(1)]
```

```python
new_frame
```

```python
new_frame.to_csv("data_2019.csv")
```
